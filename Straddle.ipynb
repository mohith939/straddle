```python
import numpy as np
```

```python
strike_price = 100   # Strike price for both call and put options
premium_call = 5     # Premium paid for the call option
premium_put = 5      # Premium paid for the put option

stock_prices = np.arange(50, 151, 1)  # Range of stock prices at expiration
```

```python
# Payoff for the call option
call_payoff = np.maximum(stock_prices - strike_price, 0) - premium_call

# Payoff for the put option
put_payoff = np.maximum(strike_price - stock_prices, 0) - premium_put

# Total payoff for the straddle strategy
total_payoff = call_payoff + put_payoff
```

```python
plt.figure(figsize=(10, 6))

# Plot the individual payoffs and the total payoff
plt.plot(stock_prices, call_payoff, label='Call Payoff', color='blue')
plt.plot(stock_prices, put_payoff, label='Put Payoff', color='green')
plt.plot(stock_prices, total_payoff, label='Total Payoff', color='red')

# Add titles and labels
plt.title('Straddle Strategy Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
